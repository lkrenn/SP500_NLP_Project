Let the weight of our portfolio/assets be $w_i$ We want to optimize our predictors subject to the following constraints:

* Each asset has a weight of less than 3\%: $|w_i| < 0.03 ~\forall~ i$
* The absolute values of our weights need to sum to our leverage: $\sum_i |w_i| = 1$
* The weights need to sum to zero: $\sum_i w_i = 0$

We can put these into the cvxpy package and have it solve for our weights:

```python
n = len(sample_predictor.columns)
w = cp.Variable(n)

# Monthly slice
p = sample_predictor.iloc[100,:].copy()
gamma = cp.Parameter(nonneg=True)
omega = sample_rets.cov().fillna(0).values
```

```python
risk = cp.quad_form(w, omega)
prob = cp.Problem(cp.Maximize(w@p.T - gamma*risk), 
               [cp.sum(w) == 0, 
                w <= 0.03,
                w >= -0.03,
               cp.sum(cp.abs(w)) <=5]) 
```

```python
# Compute trade-off curve for each gamma
SAMPLES = 10
i=5
gamma_vals = np.logspace(-2, 3, num=SAMPLES)
gamma.value = gamma_vals[i]
prob.solve()
len(w.value)
```

```python
year_month_list = [[x,y] for x in range(2006,2020) for y in range(1,13)]
year_month_list += [[2021,y] for y in range(1,7)] 
```

```python
# Initialize weight as equal weight #ToDo
w = [1/983]*983
wlist = []

for i in range(1,len(year_month_list)):
    print(i)

    date = year_month_list[i]
    prev_date = year_month_list[i-1]

    # For each month set up the parameters
    n = 983

    # Monthly slice
    p = monthly_predictor.loc[str(date[0])+'-'+str(date[1]),:].copy().fillna(0)
    gamma = cp.Parameter(nonneg=True)

    # Get Covariance matrix for previous month
    omega = (w * rets.loc[str(prev_date[0])+'-'+str(prev_date[1]),:]).copy().cov().fillna(0)#.values

    # Reinitialize weights to solve for this month
    w = cp.Variable(n)

    risk = cp.quad_form(w, omega)
    prob = cp.Problem(cp.Maximize(w@p.T - gamma*risk), 
                   [cp.sum(w) == 0, 
                    w <= 0.03,
                    w >= -0.03,
                   cp.sum(cp.abs(w)) <=5]) 

    gamma_vals = np.logspace(-2, 3, num=SAMPLES)
    gamma.value = gamma_vals[1]
    prob.solve()
    w = w.value

    wlist.append(w)
```
