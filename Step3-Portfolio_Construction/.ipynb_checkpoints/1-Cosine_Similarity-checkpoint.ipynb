# Cosine Similarity

In this notebook, we will be creating our portfolio based on the cosine similarity score. 

```python
import numpy as np
import pandas as pd
import pickle
```

```python
# Import the Cosine Scores
with open('../Step2-Signal/4-link_dict.pickle', 'rb') as f:
    parsed = pickle.load(f)
```

```python
# Flatten into a single df
df = pd.concat(parsed, axis=0).reset_index()
df.head()
```

```python
df.columns = ['company', 'doc_idx'] + list(df.columns[2:])
df.head()
```

```python
df['filingDate'] = pd.to_datetime(df['filingDate'])
```

```python
df = df.sort_values(['filingDate'])
```

```python
# Rank all the scores for a given year
scores = df[df['filingDate'].dt.year == 2020]
scores
```

```python
scores = scores.dropna(axis=0)
scores
```

```python
scores.sort_values('Cos_Sim', ascending=False)
```
